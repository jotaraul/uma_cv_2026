# Appendix 3. Visualization of kernel derivatives

This notebook defines a function called ``generalKernel()`` that, given a number, returns a kernel for computing vertical and horizontal derivatives, and another one ``toKernel()`` that translates a list of numbers into a kernel with given dimensions.

Both functions are used to create an illustrate different popular kernels.

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['figure.figsize'] = (15.0, 15.0)
```

```python
def toKernel(list, dimensions):
    matrix = np.array(list)
    matrix.resize(dimensions)
    return matrix

def generalKernel3x3(k):
    return (toKernel([1,0,-1,k,0,-k,1,0,-1],(3,3))/(2+k),toKernel([-1,-k,-1,0,0,0,1,k,1],(3,3))/(2+k))

# Definimos los kernels
ROBERTS_H = toKernel([0,0,0,0,0,1,0,-1,0], (3,3))
ROBERTS_V = toKernel([-1,0,0,0,1,0,0,0,0], (3,3))

PREWITT_H, PREWITT_V = generalKernel3x3(1)

SOBEL_H, SOBEL_V = generalKernel3x3(2)

FREI_CHEN_H, FREI_CHEN_V = generalKernel3x3(np.sqrt(2))

plt.subplot(421)
plt.imshow(ROBERTS_H, cmap='gray')
plt.title("Roberts Horizotal")
plt.subplot(422)
plt.imshow(ROBERTS_V, cmap='gray')
plt.title("Roberts Vertical")
plt.subplot(423)
plt.imshow(PREWITT_H, cmap='gray')
plt.title("Prewitt Horizontal")
plt.subplot(424)
plt.imshow(PREWITT_V, cmap='gray')
plt.title("Prewitt Vertical")
plt.subplot(425)
plt.imshow(SOBEL_H, cmap='gray')
plt.title("Sobel Horizontal")
plt.subplot(426)
plt.imshow(SOBEL_V, cmap='gray')
plt.title("Sobel Vertical")
plt.subplot(427)
plt.imshow(FREI_CHEN_H, cmap='gray')
plt.title("Frei-Chen Horizontal")
plt.subplot(428)
plt.imshow(FREI_CHEN_V, cmap='gray')
plt.title("Frei-Chen Vertical")
plt.show()
```
